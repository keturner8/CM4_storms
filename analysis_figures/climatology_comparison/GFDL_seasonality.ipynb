```python
def seasonal_avg(expname, v):
    ds = xr.open_dataset(f'{path}/clim_{expname}_{v}.nc', use_cftime=True)
    
    ds_grid = xr.open_dataset(grid[expname])
    ocean_area = ds_grid.areacello * ds_grid.wet
    SO_area = ocean_area.where(ds_grid.geolat < -40).sum(dim={"xh", "yh"})
    
    SO_total = (ds[f"dailymean_{v}"]*ocean_area).where(ds_grid.geolat < -40).sum(dim={"xh", "yh"})
    SO_avg = SO_total / SO_area
    seasonal_means = ds[f'dailymean_{v}'].groupby(ds.time.dt.season).mean()
     
    return SO_total, SO_avg, seasonal_means
```
